```python
"""Importing library"""
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

## Processing Normal X-ray Image

```python
#Load the normal X-ray image
normal_img = Image.open("normal.jpeg")

#Inspect basic properties
print(normal_img)
print("Mode: ",normal_img.mode)
```

```python
# Convert image to NumPy array (pixel matrix)
normal_array = np.array(normal_img)

# Inspect data type and intensity range
print("dtype:", normal_array.dtype)
print("min:", normal_array.min())
print("max:", normal_array.max())
```

```python
# Apply thresholding to highlight high-intensity regions
threshold = np.where(normal_array < 150, 0, 255)
```

```python
# Display the thresholded image
plt.imshow(threshold, cmap='gray')
# Add context for what the visualization represents
plt.title("Thresholded Image (Bones Only)")
# Hide axis for cleaner, image-like display
plt.axis("off")
#Render the plot
plt.show()
```

## Processing Pneumonia X-ray Image


```python
#Load the pneumonia X-ray image
pneumonia_img = Image.open("pneumonia.jpeg")

#Inspect basic properties
print(pneumonia_img)        
print("Mode: ",pneumonia_img.mode)
```

```python
#Convert pneumonia image to NumPy array(pixel matrix)
pneumonia_array = np.array(pneumonia_img)

#Inspect data type and intensity range
print("dtype:", pneumonia_array.dtype)  
print("min:", pneumonia_array.min())
print("max:", pneumonia_array.max())    
```

```python
#Apply same threshold to pneumonia image for better comparison
pneumonia_threshold = np.where(pneumonia_array < 150, 0, 255)
```

```python
#Display the thresholded pneumonia image
plt.imshow(pneumonia_threshold, cmap='gray')    
plt.title("Thresholded Pneumonia Image")
plt.axis("off")
plt.show()
```
