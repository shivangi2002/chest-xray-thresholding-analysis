```python
"""Importing library"""
import numpy as np
from PIL import Image
```

```python
"""Load the normal X-ray image"""
normal_img = Image.open("normal.jpeg")
print(normal_img)
print(normal_img.mode)
```

```python
# Convert image to NumPy array (matrix)
"""This transforms the image from a PIL object into a matrix of pixel values"""
normal_array = np.array(normal_img)

# Check data type of pixel values
print(normal_array.dtype)

#Find the minimum pixel value in the image
"""This represents the darkest pixel (closest to black)"""
print(normal_array.min())

"""Find the maximum pixel value in the image.
This represents the brightest pixel (closest to white)"""
print(normal_array.max())
```
